```python
#Q1. Program with Student, Course, and Department — Enroll Student
class Course:
    def __init__(self, name):
        self.name = name

class Department:
    def __init__(self, name):
        self.name = name
        self.courses = []

    def add_course(self, cname):
        self.courses.append(Course(cname))

class Student:
    def __init__(self, name):
        self.name = name

    def enroll(self, department, course):
        print(f"{self.name} enrolled in {course.name} of {department.name} department.")

# Example usage
dept = Department("Artificial intelligence (AI)")
dept.add_course("OOP")
dept.add_course("Data Structures")

course = Course("PYTHON")
student = Student("Rohit")

student.enroll(dept, course)
```

```python
#Q2. Program with Bill—Pay by Cash or Cheque Using Inheritance
class Bill:
    def __init__(self, amount):
        self.amount = amount

    def pay(self):
        pass  # overridden in subclasses

class Cash(Bill):
    def pay(self):
        print(f"Paid ₹{self.amount} by Cash")

class Cheque(Bill):
    def __init__(self, amount, cheque_no):
        super().__init__(amount)
        self.cheque_no = cheque_no

    def pay(self):
        print(f"Paid ₹{self.amount} by Cheque (No: {self.cheque_no})")

# Example
b1 = Cash(500)
b2 = Cheque(1000, "CHQ1122")

b1.pay()
b2.pay()
```

```python
#Q3. Overload -= Operator for Distance Objects (Using Python Magic Method)
class Distance:
    def __init__(self, meters=0):
        self.meters = meters

    def __isub__(self, other):
        self.meters -= other.meters
        return self

# Example
d1 = Distance(50)
d2 = Distance(20)

d1 -= d2

print("Result:", d1.meters, "meters")
```

```python
#Q4. Two Classes for Distance: Meters and Kilometers + Conversions
class Meter:
    def __init__(self, m=0):
        self.m = m

class Kilometer:
    def __init__(self, km=0):
        self.km = km

def to_km(distance_meter):
    return Kilometer(distance_meter.m / 1000)

def to_m(distance_km):
    return Meter(distance_km.km * 1000)

# Example
m1 = Meter(2500)
k1 = Kilometer(2)

converted_km = to_km(m1)
converted_m = to_m(k1)

print("2500 meters =", converted_km.km, "km")
print("2 km =", converted_m.m, "meters")
```
