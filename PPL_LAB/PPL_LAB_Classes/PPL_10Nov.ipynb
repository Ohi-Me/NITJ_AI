```python
n=int(input("Size of dictionary : "))
dict={}
for i in range(n):
    x=int(input("enter size of element in that index dict : "))
    l=[]
    for j in range(x):
        num=input(f"Enter value for {j}th key : ")
        l.append(num)
    dict[(i+1)]=l

for i in range(n):
    for j in range(n):
        if (i!=j):
            for a in dict[i+1]:
                for b in dict[j+1]:
                    print(a+b)
```

```python
d={'a':5,'b':14,'c':32,'d':35,'e':24,'f':100,'g':57,'h':8,'i':100}
vals=[]
for v in d.values():
 if v not in vals:
  vals.append(v)
for i in range(len(vals)):
 for j in range(i+1,len(vals)):
  if vals[i]<vals[j]:
   t=vals[i]
   vals[i]=vals[j]
   vals[j]=t
for n in [1,2,5]:
 top=vals[:n]
 print(n,'maximum value(s) in the said dictionary:')
 out=[]
 for k in d:
  if d[k] in top:
   out.append(k)
 print(out)
```
