```python
#wap simulate  and analyze traffic movement on a single road with several lanes using lists
#genrate random vehicles
#simulate entry and exit 
#measure congestion flow and average wait time

import random

LANES = 4
CAPACITY = 5
TIME = 20
ENTRY_PROB = 0.5
EXIT_PROB = 0.4
# random.seed(42)

lanes = [[random.randint(1, 100) for _ in range(random.randint(3, CAPACITY - 2))] for _ in range(LANES)]
entry_time, exit_times = {}, []
vid = 101

print("Initial Lanes:")
for i, lane in enumerate(lanes, 1):
    print(f"Lane {i}: {lane}")

for t in range(1, TIME + 1):
    for lane in lanes:
        # Exit car
        if lane and random.random() < EXIT_PROB:
            car = lane.pop()
            wait = t - entry_time.get(car, t)
            exit_times.append(wait)
        # Enter car
        if len(lane) < CAPACITY and random.random() < ENTRY_PROB:
            lane.insert(0, vid)
            entry_time[vid] = t
            vid += 1

if exit_times:
    total_wait = 0
    for w in exit_times:
        total_wait += w
    avg_wait = total_wait / len(exit_times)
else:
    avg_wait = 0

congestion = []
for lane in lanes:
    congestion.append((len(lane) / CAPACITY) * 100)

flow = len(exit_times) / TIME

print("Average Wait Time :", round(avg_wait, 2))
print("Flow Rate         :", round(flow, 2), "/step")
print("Lane Congestions  :", [str(round(c, 1)) + "%" for c in congestion])

```
