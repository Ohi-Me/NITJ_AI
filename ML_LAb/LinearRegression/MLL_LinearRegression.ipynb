```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
x = np.random.rand(50) * 10   # 50 random values between 0â€“10
y = 3*x + 5 + np.random.randn(50)*2   # true relation + noise
```

```python
plt.scatter(x, y, color='blue')
plt.title("Generated Data (y = 3x + 5 + noise)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
X = np.c_[np.ones(x.shape[0]), x]  # shape (50,2)
Y = y.reshape(-1,1)
```

```python
theta = np.linalg.inv(X.T @ X) @ X.T @ Y
print("Coefficients (intercept, slope):", theta.ravel())
```

```python
y_pred = X @ theta
```

```python
plt.scatter(x, y, color='blue', label='Actual data')
plt.plot(x, y_pred, color='red', label='Predicted line')
plt.title("Linear Regression Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
mse = np.mean((y - y_pred.ravel())**2)
ss_res = np.sum((y - y_pred.ravel())**2)
ss_tot = np.sum((y - np.mean(y))**2)
r2 = 1 - (ss_res/ss_tot)

print("Mean Squared Error:", mse)
print("RÂ² Score:", r2)
```
