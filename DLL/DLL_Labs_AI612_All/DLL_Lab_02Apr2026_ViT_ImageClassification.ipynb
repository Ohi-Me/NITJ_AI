```python
import torch, numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import timm
from torchvision import transforms
from torchvision.datasets import CIFAR10
from torch.utils.data import DataLoader
import torch.nn as nn
import warnings; warnings.filterwarnings('ignore')
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}  |  timm: {timm.__version__}')
```

## Load ViT-B/16

```python
model = timm.create_model('vit_base_patch16_224', pretrained=True, num_classes=10)
model = model.to(device)
total = sum(p.numel() for p in model.parameters())
print(f'Model         : ViT-B/16')
print(f'Parameters    : {total:,}')
print(f'Image size    : 224x224')
print(f'Patch size    : 16x16  -> 196 patches + 1 CLS = 197 tokens')
print(f'Embed dim     : 768  |  Heads: 12  |  Layers: 12')
print(f'Output classes: 10 (CIFAR-10 fine-tune)')
```

## Fine-tuning History (Simulated)

```python
epochs = 10
train_acc = [0.6823,0.8012,0.8534,0.8812,0.8978,0.9089,0.9167,0.9234,0.9289,0.9334]
val_acc   = [0.6534,0.7823,0.8312,0.8589,0.8756,0.8867,0.8945,0.9012,0.9067,0.9112]
train_loss= [1.1234,0.7823,0.5934,0.4812,0.4023,0.3456,0.3012,0.2678,0.2412,0.2198]
val_loss  = [1.2345,0.8234,0.6123,0.5012,0.4234,0.3678,0.3234,0.2912,0.2656,0.2456]
print(f"{'Epoch':>6}  {'Train Acc':>10}  {'Val Acc':>9}  {'Val Loss':>9}")
print('-'*40)
for i in range(epochs):
    print(f'{i+1:>6}  {train_acc[i]:>10.4f}  {val_acc[i]:>9.4f}  {val_loss[i]:>9.4f}')
print(f'\nBest Val Accuracy: {max(val_acc)*100:.2f}%')
```

```python
ep = list(range(1, epochs+1))
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(ep, train_acc, label='Train', linewidth=2)
axes[0].plot(ep, val_acc, label='Validation', linewidth=2)
axes[0].set_title('ViT-B/16 Accuracy (CIFAR-10)')
axes[0].set_xlabel('Epoch'); axes[0].set_ylabel('Accuracy')
axes[0].set_ylim([0.5, 1.0]); axes[0].legend(); axes[0].grid(True, alpha=0.3)

axes[1].plot(ep, train_loss, label='Train', linewidth=2)
axes[1].plot(ep, val_loss, label='Validation', linewidth=2)
axes[1].set_title('ViT-B/16 Loss (CIFAR-10)')
axes[1].set_xlabel('Epoch'); axes[1].set_ylabel('Loss')
axes[1].legend(); axes[1].grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
```
