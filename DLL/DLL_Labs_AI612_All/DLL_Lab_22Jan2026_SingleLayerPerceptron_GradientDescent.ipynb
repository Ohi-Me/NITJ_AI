## Single Layer Perceptron with Gradient Descent
**Done By:** Student | **Course:** AI-612


```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings('ignore')
print('Libraries loaded.')
```

## Define Model

```python
w = tf.Variable([.2], dtype=tf.float32)
b = tf.Variable([-.2], dtype=tf.float32)
x = tf.placeholder(tf.float32, name='x')
y = tf.placeholder(tf.float32, name='y')
linear_model = w * x + b
print('Model: y_hat = w*x + b')
print('Initial w:', [0.2], '  Initial b:', [-0.2])
```

## Loss and Optimizer

```python
squared_deltas = tf.square(linear_model - y)
loss = tf.reduce_sum(squared_deltas)
optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(loss)
print('Loss: Sum of Squared Errors')
print('Optimizer: Gradient Descent  lr=0.01')
```

## Train for 1000 Epochs

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

x_data = [1, 2, 3, 4]
y_data = [0, -1, -2, -3]

loss_before = sess.run(loss, {x: x_data, y: y_data})
print(f'Loss before training: {loss_before:.4f}')

loss_history = []
for i in range(1000):
    sess.run(train, {x: x_data, y: y_data})
    loss_history.append(sess.run(loss, {x: x_data, y: y_data}))

loss_after = sess.run(loss, {x: x_data, y: y_data})
final_w, final_b = sess.run([w, b])
print(f'Loss after  training: {loss_after:.2e}')
print(f'Final w: {final_w}  Final b: {final_b}')
```

## Predictions

```python
preds = sess.run(linear_model, {x: x_data, y: y_data})
print(f'{'x':>4}  {'y_true':>8}  {'y_pred':>10}')
print('-'*26)
for xi, yi, yp in zip(x_data, y_data, preds):
    print(f'{xi:>4}  {yi:>8}  {yp:>10.5f}')
```

## Loss Curve

```python
plt.figure(figsize=(7, 4))
plt.plot(loss_history, color='steelblue', linewidth=2)
plt.yscale('log')
plt.title('Training Loss over 1000 Epochs', fontsize=13)
plt.xlabel('Epoch'); plt.ylabel('Loss (log scale)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
print('Loss curve plotted.')
```

## Summary

Gradient Descent successfully converged the perceptron to the true relationship **y = -x + 1**.
