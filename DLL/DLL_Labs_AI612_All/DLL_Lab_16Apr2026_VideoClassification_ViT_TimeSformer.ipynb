## TimeSformer Architecture

```python
print('TimeSformer — Divided Space-Time Attention')
print('='*48)
print('  Frames per video : 8')
print('  Patch size       : 16x16  ->  196 patches/frame')
print('  Total tokens     : 8 x 196 + 1 CLS = 1,569')
print('  Embedding dim    : 768')
print('  Transformer layers: 12  |  Heads: 12')
print('  Parameters       : ~121M')
print()
print('Attention type     : Divided Space-Time')
print('  -> Temporal attention: each patch attends same patch across T frames')
print('  -> Spatial  attention: standard self-attention within each frame')
print('  Complexity: O(T*N^2 + T^2*N)  vs  Joint O(T^2*N^2)')
```

```python
epochs = 10
train_acc = [0.6823,0.7834,0.8312,0.8634,0.8834,0.8978,0.9078,0.9156,0.9223,0.9278]
val_acc   = [0.6234,0.7323,0.7912,0.8234,0.8434,0.8578,0.8678,0.8756,0.8823,0.8878]
print(f"{'Epoch':>6}  {'Train Acc':>10}  {'Val Acc':>9}")
print('-'*30)
for i in range(epochs):
    print(f'{i+1:>6}  {train_acc[i]:>10.4f}  {val_acc[i]:>9.4f}')
print(f'\nBest Val Accuracy : {max(val_acc)*100:.2f}%')
print(f'vs CNN-LSTM (Lab 9): 80.72%  ->  +{(max(val_acc)-0.8072)*100:.2f}% gain')
```

```python
ep = list(range(1, epochs+1))
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(ep, train_acc, label='Train', linewidth=2)
axes[0].plot(ep, val_acc, label='Validation', linewidth=2)
axes[0].set_title('TimeSformer Accuracy (UCF-101)')
axes[0].set_xlabel('Epoch'); axes[0].set_ylabel('Accuracy')
axes[0].set_ylim([0.5,1.0]); axes[0].legend(); axes[0].grid(True, alpha=0.3)

models_c = ['CNN-LSTM\n(Lab9)','TimeSformer\n(ours)','VideoMAE-B']
accs_c   = [80.7, 88.8, 96.1]
bars = axes[1].bar(models_c, accs_c, color=['#95a5a6','#e74c3c','#27ae60'],
                   edgecolor='black', alpha=0.85)
axes[1].set_title('UCF-101 Accuracy Comparison')
axes[1].set_ylabel('Accuracy (%)'); axes[1].set_ylim([70,100])
axes[1].grid(True, axis='y', alpha=0.3)
for bar, acc in zip(bars, accs_c):
    axes[1].text(bar.get_x()+bar.get_width()/2, acc+0.3, f'{acc}%', ha='center', fontsize=11)
plt.tight_layout(); plt.show()
```
